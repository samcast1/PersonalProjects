```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
bstud_data = pd.read_csv('map.csv')
bstud_data.columns = bstud_data.loc[0]
bstud_data.drop(0, inplace = True)
bstud_data['name'] = bstud_data['first'] + " " + bstud_data['last']
bstud_data.drop(columns = ['first','last'], inplace = True)
bstud_data[['Latitude','Longitude']] = bstud_data.coord.str.split(', ', expand = True)
bstud_data.Longitude = bstud_data.Longitude.astype(str)
bstud_data.Longitude = bstud_data.Longitude.apply(lambda x: x.replace(" ", ""))
bstud_data.Longitude = bstud_data.Longitude.astype(float)
bstud_data.Latitude = bstud_data.Latitude.astype(float)
bstud_data.cur_bstud_lead.fillna('N', inplace = True)
bstud_data.focus_inv.fillna('None', inplace = True)
bstud_data.drop(29, inplace = True)
bstud_data.info()
```

```python
bstud_list = []
for index, row in bstud_data.iterrows() :
    if row['cur_bstud_lead'] == 'Yes!':
        bstud_list.append('Current Leader')
    elif row['cur_bstud_lead'] == 'I would like to start one!' or 'Lead a Bible Study' in row['activity_interest']:
        bstud_list.append('New Leader')
    elif row['cur_bstud_lead'] == 'N' and 'Lead a Bible Study' not in row['activity_interest'] and 'Join a Bible Study' in row['activity_interest'] :
        bstud_list.append('New Attendee')
   # elif row['cur_bstud_lead'] == 'N' and 'Lead a Bible Study' in row['activity_interest'] :
    #    bstud_list.append('New Leader')
    #elif row['cur_bstud_lead'] == 'N' and 'Join a Bible Study' not in row['activity_interest'] or 'Lead a Bible Study' not in row['activity_interest'] :
     #   bstud_list.append('No Bstud Interest')
    else :
        bstud_list.append('No Bstud Interest')
bstud_data['bstud_status'] = bstud_list
```

```python
serve_list = []
for index, row in bstud_data.iterrows() :
    if 'Serve Local Community' in row['activity_interest'] :
        serve_list.append('Interested in Serving Homeless')
    else :
        serve_list.append('Not Interested')
bstud_data['serve_status'] = serve_list
```

```python
bstud_data.focus_inv.unique()
```

```python
focus_list = []
for index, row in bstud_data.iterrows() :
    if 'Full-Time Missionary' in row['focus_inv'] :
        focus_list.append('Missionary Alum')
    elif 'Full-Time Missionary' not in row['focus_inv'] and 'Bible' in row['focus_inv'] :
        focus_list.append('Student Alum')
    else :
        focus_list.append('Not Alum')
bstud_data['focus_status'] = focus_list
```

```python
new_map = folium.Map(location=[32.7767, -96.7970], zoom_start=10)
color_map_male = {
    'Current Leader' : 'lightblue',
    'New Leader' : 'green',
    'New Attendee' : 'orange',
    'No Bstud Interest' : 'black'
}
color_map_female = {
    'Current Leader' : 'pink',
    'New Leader' : 'yellow',
    'New Attendee' : 'red',
    'No Bstud Interest' : 'gray'
}

marker_cluster = MarkerCluster().add_to(new_map)
default_color = 'gray'

#icon_male = folium.Icon(color='blue', icon='male', prefix='fa')
#icon_female = folium.Icon(color='pink', icon='female', prefix='fa')

legend_html = '''
<div style="position: fixed; 
            bottom: 10px; left: 10px; width: 150px; 
            background-color: white; border:2px solid grey; z-index:9999; font-size:12px; padding: 5px;">
    <p style="margin-bottom: 3px;"><b>Legend</b></p>'''


# Add legend entries for each color in the color maps
for status, color in color_map_male.items():
    legend_html += f'<p style="margin: 2px;"><i class="fa fa-map-marker fa-2x" style="color:{color};"></i> Male: {status}</p>'

for status, color in color_map_female.items():
    legend_html += f'<p style="margin: 2px;"><i class="fa fa-map-marker fa-2x" style="color:{color};"></i> Female: {status}</p>'

# Close the legend HTML
legend_html += '</div>'

# Add legend to the map
new_map.get_root().html.add_child(folium.Element(legend_html))

for index, row in bstud_data.iterrows():
    coordinates = (row['Latitude'], row['Longitude'])
    characteristic_value = {
        'Name': row['name'],
        'Parish': row['parish'],
        'Current Faith': row['cur_faith'],
        'Faith Goals': row['grow_faith'],
        'Activity Interests': row['activity_interest'],
        'Homeless Ministry': row['serve_status'],
        'FOCUS Involvement' : row['focus_status'],
        'Additional Notes': row['add_notes']        
    }
    popup_content = "<br>".join([f"<b>{key}:</b> {value}" for key, value in characteristic_value.items()])
    popup_style = "width: 300px;"
    if row['gender'] == 'Male':
        icon_color = color_map_male.get(row['bstud_status'], default_color)
        #icon = icon_male
    elif row['gender'] == 'Female':
        icon_color = color_map_female.get(row['bstud_status'], default_color)
        #icon = folium.Icon(color='pink', icon='female', prefix='fa')
    else:
        icon_color = default_color 
        #icon = folium.Icon(color=icon_color)
    folium.Marker(location=coordinates, popup=folium.Popup(popup_content, max_width=300, min_width=250, style=popup_style), icon=folium.Icon(color=icon_color)).add_to(marker_cluster)
```
